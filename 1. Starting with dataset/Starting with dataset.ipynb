```python
# выбрать только те дни, которые выпали на праздники (праздников не так много)
bikes['Holiday'] == 'Holiday'
```

```python
# использовать полученный ранее массив для фильтрации
bikes[bikes['Holiday'] == 'Holiday']
```

```python
# получить праздничные дни с температурой >=15 градусов
bikes[(bikes['Holiday'] == 'Holiday') & (bikes['Temperature'] >= 15)]
```

```python
# получить или праздники или дни с температурой >= 15
bikes[(bikes['Holiday'] == 'Holiday') | (bikes['Temperature'] >= 15)].shape
```

```python
# получить не праздники и не жарче 15 градусов - используется тильда ~
bikes[~((bikes['Holiday'] == 'Holiday') | (bikes['Temperature'] >= 15))].shape
```

```python
# проверяет является ли ячейка без значения (так же можно использовать как условие для фильтрации)
bikes['Temperature'].isna()
```

```python
# фильтрация по .isna()
bikes[bikes['Temperature'].isna()]
```

```python
# соединение условий с помощью is
bikes['Seasons'].isin(['Winter', 'Summer'])
```

```python
# использование полученного через isin массива в качестве фильтра
bikes[bikes['Seasons'].isin(['Winter', 'Summer'])].shape
```

```python
# в isin так же можно передавать интервал
bikes[bikes['Temperature'].isin(range(15, 26))]
```
